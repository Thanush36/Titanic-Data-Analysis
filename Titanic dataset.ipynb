```python
# Selecting numerical columns for the box plot
numerical_columns = df_train.select_dtypes(include=['int64', 'float64']).columns

# Creating a box plot to visualize outliers in numerical columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_train[numerical_columns])
plt.xticks(rotation=45)
plt.title("Box Plot: Outliers Detection")
plt.xlabel("Columns")
plt.ylabel("Values")
plt.show()
```

## Removing Outliers in 'Age' and 'Fare' columns

```python
# Function to remove outliers based on IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Removing outliers from 'Age' and 'Fare' columns
df_train_no_outliers = remove_outliers(df_train, 'Age')
df_train_no_outliers = remove_outliers(df_train_no_outliers, 'Fare')

# Displaying the shape of the DataFrame after removing outliers
print("Shape of DataFrame after removing outliers:", df_train_no_outliers.shape)
```
